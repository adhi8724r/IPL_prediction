```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import pickle
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
df=pd.read_csv("Dataset//final_ipl_data.csv")
```

```python
team_mapping = {
    'Royal Challengers Bangalore': 'RCB',
    'Delhi Daredevils': 'DC',
    'Kolkata Knight Riders': 'KKR',
    'Mumbai Indians': 'MI',
    'Rajasthan Royals': 'RR',
    'Chennai Super Kings': 'CSK',
    'Deccan Chargers': 'SRH',
    'Sunrisers Hyderabad': 'SRH',
    'Delhi Capitals': 'DC',
    'Punjab Kings': 'PBKS',
    'Gujarat Titans': 'GT',
    'Lucknow Super Giants': 'LSG',
    'Royal Challengers Bengaluru': 'RCB'
}
```

```python
df['batting_team']=df['batting_team'].replace(team_mapping)
df['bowling_team']=df['bowling_team'].replace(team_mapping)
```

```python
df['batting_team'].unique()
```

```python
df['city'].unique()
```

```python
df.head(500)
```

```python
Le_team=LabelEncoder()
df['batting_team']=Le_team.fit_transform(df['batting_team'])
df['bowling_team']=Le_team.transform(df['bowling_team'])
```

```python
df.head(500)
```

```python
Le_city=LabelEncoder()
df['city']=Le_city.fit_transform(df['city'])
```
